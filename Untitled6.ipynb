```python
# Rishabh Thakur - 21BDS0183
print("Data Summarization, Visualization, and Clustering Analysis - Rishabh Thakur (21BDS0183)")

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, KBinsDiscretizer, StandardScaler
from sklearn.cluster import SpectralClustering, KMeans
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.metrics import silhouette_score

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bcancer.csv"
df = pd.read_csv(url)

# Drop empty and irrelevant columns
df.drop(columns=['Unnamed: 32'], inplace=True, errors='ignore')

# Display basic info
print("\nDataset Info:")
print(df.info())

# Display first few rows
print("\nFirst 5 Rows:")
print(df.head())

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Encoding categorical columns using Label Encoding
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Check for duplicates and remove them
print("\nNumber of Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()

# Check for missing values before handling
print("\nMissing Values Before Handling:\n", df.isnull().sum())

# Handling Missing Data using Mean Imputation
imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# Check missing values after imputation
print("\nMissing Values After Imputation:\n", df.isnull().sum())

# Discretization (Binning) of a numerical column
if len(numerical_cols) > 0:
    binning = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
    df[numerical_cols[0] + "_binned"] = binning.fit_transform(df[[numerical_cols[0]]])

# **1D Statistical Analysis**
print("\n1D Statistical Analysis:")
print(df.describe())

# **2D Statistical Analysis** (Correlation Matrix)
print("\n2D Statistical Analysis:")
corr_matrix = df.corr()
print(corr_matrix)

# **nD Statistical Analysis** (Pairwise correlation of top 5 features)
print("\nnD Statistical Analysis:")
top_features = corr_matrix.abs().sum().sort_values(ascending=False).index[:5]
print(df[top_features].corr())

# **Contingency Table**
if len(categorical_cols) > 1:
    contingency_table = pd.crosstab(df[categorical_cols[0]], df[categorical_cols[1]])
    print("\nContingency Table:")
    print(contingency_table)

# **Visualization - Scatter plots, Dot charts, Bar plots**
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df[numerical_cols[0]], y=df[numerical_cols[1]], hue=df[categorical_cols[0]])
plt.title("Scatter Plot")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=df[categorical_cols[0]], y=df[numerical_cols[0]])
plt.title("Boxplot")
plt.show()

plt.figure(figsize=(10, 6))
df[categorical_cols[0]].value_counts().plot(kind='bar', color='skyblue')
plt.title("Bar Plot")
plt.show()

# **Clustering Analysis - Spectral Clustering**
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numerical_cols])

spectral_model = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=42)
labels = spectral_model.fit_predict(X_scaled)
df['Spectral_Cluster'] = labels
print("\nSpectral Clustering Completed.")

# **Hierarchical Clustering**
linkage_matrix = linkage(X_scaled, method='ward')
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# **Outlier Detection using Clustering (KMeans)**
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
df['KMeans_Cluster'] = kmeans_labels
silhouette_avg = silhouette_score(X_scaled, kmeans_labels)
print(f"\nSilhouette Score for KMeans: {silhouette_avg:.4f}")

# Save cleaned dataset
df.to_csv("cleaned_bcancer_with_clustering.csv", index=False)
print("\n Data Processing, Summarization, Visualization, and Clustering Completed. Cleaned dataset saved as 'cleaned_bcancer_with_clustering.csv'")
```
